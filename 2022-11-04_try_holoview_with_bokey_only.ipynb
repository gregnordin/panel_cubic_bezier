# Try Holoviews with Bokeh only

[Holoviews - Plotting with Bokeh](https://holoviews.org/user_guide/Plotting_with_Bokeh.html#sizing-elements)

```python
xs = np.linspace(0, np.pi*4, 100)
data = (xs, np.sin(xs))
data2 = (xs, np.cos(xs))

points_a = hv.Points(data)
points_b = hv.Points(data2)

points_a.opts(width=300, height=300) + points_b.opts(width=600, height=300, color="red")
```

```python
points_a.opts(width=300, height=300) * points_b.opts(width=600, height=300, color="red")
```
